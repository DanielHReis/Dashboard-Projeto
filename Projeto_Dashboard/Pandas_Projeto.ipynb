```python
df.drop(columns=['Unnamed: 0'], inplace=True)
df.drop(columns=['salary_currency'], inplace=True)
df.drop(columns=['salary'], inplace=True)
```

```python
df.rename(columns={
    'work_year': 'Ano de Referencia',
    'experience_level': 'Nivel de Experiencia',
    'salary_in_usd': 'Salario em Dolar(Anual)',
    'employment_type': 'Tipo de Contratacao',
    'job_title': 'Cargo',
    'employee_residence': 'Residencia Funcionario',
    'remote_ratio': 'Modalidade de Trabalho',
    'company_location': 'Residencia Empresa',
    'company_size': 'Tamanho da Empresa'
    }, inplace=True)
```

```python
df.head()
```

```python
experiencia = {
    'EN': 'Junior',
    'MI': 'Pleno',
    'SE': 'Senior',
    'EX': 'Executivo'
}

df['Nivel de Experiencia'] = df['Nivel de Experiencia'].replace(experiencia)
```

```python
contratacao = {
    'FT': 'Tempo Integral',
    'PT': 'Meio Periodo',
    'FL': 'FreeLancer',
    'CT': 'Contrato(PJ)'
}

df['Tipo de Contratacao'] = df['Tipo de Contratacao'].replace(contratacao)
```

```python
modalidade = {
    0: 'Presencial',
    50: 'Hibrido',
    100: 'Online'
}

df['Modalidade de Trabalho'] = df['Modalidade de Trabalho'].replace(modalidade)
```

```python
df['Modalidade de Trabalho'] = df['Modalidade de Trabalho'].astype(str)
```

```python
df['Modalidade de Trabalho'].dtype
```

```python
tamanho = {
    'L': 'Grande',
    'S': 'Pequena',
    'M': 'Media'
}

df['Tamanho da Empresa'] = df['Tamanho da Empresa'].replace(tamanho)
```

### Análise de Outliers
Observamos no boxplot que existem salários acima de $400,000. Estes podem ser outliers ou cargos de alto nível (Executivos). Decidimos mantê-los pois são dados reais e importantes para a análise de mercado.

```python
fig_salario_experiencia = px.histogram(df, x='Nivel de Experiencia', title='Quantidade de Registros por Nível de Experiência')
fig_salario_experiencia.show()
```

```python
media_por_ano = df.groupby('Ano de Referencia')['Salario em Dolar(Anual)'].mean().reset_index()
fig_salario_ano = px.line(media_por_ano, x='Ano de Referencia', y='Salario em Dolar(Anual)', title='Evolução do Salário Médio Anual (USD)')
fig_salario_ano.show()
```
